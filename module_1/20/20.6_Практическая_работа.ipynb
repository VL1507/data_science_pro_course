## 0. Импорты

```python
import pandas as pd

from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

## 1. Загрузите датасет и ознакомьтесь с ним.

```python
df = pd.read_csv("./20.6_vehicles_dataset_upd.csv")
```

```python
df.head()
```

```python
df.info()
```

## 2. Сгенерируйте дополнительные признаки для этапа моделирования, используя lambda-функции и функции над строками.

```python
df['age_category'] =  df.apply(lambda x: "new" if x.year > 2013 else ("old" if x.year < 2006 else "average"), axis=1)
df['age_category']
```

```python
df["miles_per_year"] = df.apply(lambda x: x.odometer / (2025 - x.year), axis=1)
df["miles_per_year"]
```

```python
df["region_clear"] = df["region"].str.lower().str.split("/").str[0]
df["region_clear"]
```

## 3. Нормализуйте количественные признаки.

```python
std_scaler = StandardScaler()
```

```python
col_for_normal = ["price", "odometer"]
```

```python
std_scaler.fit(df[col_for_normal])
```

```python
std_scaler.mean_
```

```python
std_scaler.var_
```

```python
std_scaler.transform(df[col_for_normal])
```

```python
df[["std_scaler_price", "std_scaler_odometer"]] = std_scaler.transform(df[col_for_normal])
```

```python
df[["price", "std_scaler_price", "odometer", "std_scaler_odometer"]]
```

## 4. Примените кодирование категориальных переменных.

```python
ohe = OneHotEncoder(sparse_output=False)
```

```python
col_for_encode = ["fuel", "transmission", "price_category", "age_category"]
```

```python
ohe.fit(df[col_for_encode])
```

```python
ohe.get_feature_names_out()
```

```python
df[ohe.get_feature_names_out()] = ohe.transform(df[col_for_encode])
```

```python
df[col_for_encode + list(ohe.get_feature_names_out())]
```

## 5. Выделите признаки из временны́х переменных.

```python
import datetime as dt
```

```python
df["posting_date"] = pd.to_datetime(df["posting_date"], utc=True)
```

### Кол-во дней с момента публикации объявления

```python
df["posting_date"].info()
```

```python
df["posting_date"].info()
```

```python
df["ad_age_days"] = df["posting_date"].apply(lambda x: (dt.datetime.now(dt.timezone.utc) - x).days)
df["ad_age_days"]
```

### День недели

```python
df["posting_day_of_week"] = df["posting_date"].apply(lambda x: x.weekday())
df["posting_day_of_week"]
```

```python
df["posting_day_of_week"].value_counts().sort_index()
```
